```python
import pandas as pd
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn import linear_model,preprocessing
```

####  Load the Car Data to the Worksheet

```python
car_data = pd.read_csv("E:\DataScience\Cars\Car Data Set\car.data")
car_data.head()
```

#### We will have to convert the non numerical data into numerical data.(Using Sklearn)

```python
le  = preprocessing.LabelEncoder()
buying = le.fit_transform(list(car_data["buying"]))
maint = le.fit_transform(list(car_data["maint"]))
door = le.fit_transform (list(car_data["door"]))
persons = le.fit_transform(list(car_data["persons"]))
lug_boot = le.fit_transform(list(car_data["lug_boot"]))
safety = le.fit_transform(list(car_data["safety"]))
cls = le.fit_transform(list(car_data["class"]))
```

```python
predict = "class"
```

```python
x = list(zip(buying,maint,door,persons,lug_boot,safety))
y = list(cls)
```

```python
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(x,y, test_size = 0.1)
```

```python
print(x_train,y_test)
```

```python
model = KNeighborsClassifier(n_neighbors = 6)
model.fit (x_train,y_train)
accuracy = model.score(x_test,y_test)
print(accuracy)
```

```python
predicted = model.predict(x_test)
names = ["unacc","accuracy","good","verygood"]
for x in range (len(x_test)):
    print("Predicted: ",names[predicted[x]],"Data:",x_test[x],"Actual: ",names[y_test[x]])
```
